# Scatter Plots


```python
#Import libraries
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Loading Dataset

```python
 InsauranceData= pd.read_csv("insurance.csv")
```

```python
InsauranceData
```

# Data Examine

```python
InsauranceData.head()
```

# Scatter plots

```python
sns.scatterplot(x=InsauranceData['bmi'], y=InsauranceData['charges'])
```

```python
sns.regplot(x=InsauranceData['bmi'], y=InsauranceData['charges'])
```

# Color-coded scatter plots

```python
sns.scatterplot(x=InsauranceData['bmi'], y=InsauranceData['charges'], hue=InsauranceData['smoker'])
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=InsauranceData)
```

```python
sns.swarmplot(x=InsauranceData['smoker'],
              y=InsauranceData['charges'])
```
