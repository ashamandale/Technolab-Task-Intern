# Handling Missing Values

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('NFL Play by Play 2009-2017 (v4).csv')
```

```python
np.random.seed(0)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull()
```

```python
mv = df.isnull().sum()
```

```python
mv[0:15]
```

```python
total_cells = np.product(df.shape)
total_missing = mv.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
df.dropna()
```

```python
dropnavalue= df.dropna(axis=1)
dropnavalue.head()
```

```python
print("Columns in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % dropnavalue.shape[1])
```
